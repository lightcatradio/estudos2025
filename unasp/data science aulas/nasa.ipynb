```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('GLB.Ts+dSST.csv', skiprows=1)

df.set_index('Year', inplace=True)

df_numeric = df.apply(pd.to_numeric, errors='coerce')

media_anual = df_numeric.mean(axis=1)

variacao = media_anual.pct_change() * 100

df_decada = media_anual.groupby((media_anual.index // 10) * 10).mean()

plt.figure(figsize=(15, 5))
plt.plot(media_anual.index, media_anual.values, label='Temp. Média Anual')
plt.plot(df_decada.index, df_decada.values, color='orange', label='Temp. Média por Década')
plt.title('Temperatura Média Global')
plt.xlabel('Ano')
plt.ylabel('Temperatura (°C)')
plt.grid(True)
plt.legend()
plt.show()
```
